```python
class BankAccount:
    def __init__(self, account_holder, account_number, balance=0):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} rupees deposited successfully.")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"{amount} rupees withdrawn successfully.")
        else:
            print("Insufficient balance or invalid amount.")

    def check_balance(self):
        print(f"Current balance: {self.balance} rupees")

    def account_details(self):
        print("Account Details:")
        print(f"Holder: {self.account_holder}")
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.balance} rupees")


account1 = BankAccount("Abraz Baloch", "123456789")
account1.account_details()

account1.deposit(5000)
account1.check_balance()

account1.withdraw(2000)
account1.check_balance()
```

```python
import pandas as pd

category_keywords = {
    'Information Technology (IT)': ['developer', 'engineer', 'software', 'it', 'network', 'devops', 'technician'],
    'Data & Analytics': ['data', 'analyst', 'machine learning', 'business intelligence', 'ai', 'ml'],
    'Marketing & Sales': ['marketing', 'sales', 'seo', 'content', 'brand'],
    'Finance & Accounting': ['finance', 'accountant', 'auditor', 'tax', 'payroll'],
    'Healthcare': ['nurse', 'doctor', 'medical', 'pharmacist', 'healthcare'],
    'Education': ['teacher', 'professor', 'instructor', 'tutor', 'lecturer'],
    'Human Resources (HR)': ['hr', 'recruiter', 'talent', 'human resources'],
    'Engineering': ['civil', 'mechanical', 'electrical', 'project engineer'],
    'Customer Service': ['customer service', 'call center', 'support', 'client'],
    'Administration & Support': ['admin', 'receptionist', 'assistant', 'office']
}

def assign_category(title):
    title = str(title).lower()
    for category, keywords in category_keywords.items():
        if any(keyword in title for keyword in keywords):
            return category
    return 'Other'

df['Categories'] = df['Job Title'].apply(assign_category)
```
